## profile 

```python
profile_height = 250
```

```python
def plot_histogram(image_index):
    fig1 = plt.figure(figsize=(20, 10))

    horizontal_profile = image_data[image_index][profile_height,:]
    
    ax_img1 = plt.subplot(121)
    ax_img1.plot(horizontal_profile, '.')
    ax_img1.set_xlabel('Pixel')
    ax_img1.set_ylabel("Number of pixels")

    ax_img2 = plt.subplot(122)
    ax_img2.imshow(image_data[image_index], cmap='viridis')
    ax_img2.axhline(profile_height, color='w')
    plt.show()
    
preview = interact(plot_histogram,
              image_index = widgets.IntSlider(min=0, 
                                            max=len(image_data)-1,
                                            step=1,
                                            value=0,
                                            description='Image Index',
                                            continuous_update=False))



```

## mid max-min 

this algorithm will 
* calculate the average of the first and last plateau
* min value array
* which pixel is closer before and after the min value pixel to the mean of \[plateau,min value\]. To do this we will start at the last point of the plateau (for the first plateau) and first point of the plateau (for the last plateau) and will make our way toward the center and first pixel that is below mid range of plateau mean and min value will be our side!


```python
first_plateau_range = [0, 50]
last_plateau_range = [360, len(image_data[0][profile_height,:])]
```

```python
## use for debugging to figure out plateau range

# fig3 = plt.figure(figsize=(20, 10))

# image_index = 0
# horizontal_profile = image_data[0][profile_height,:]

# ax_img1 = plt.subplot(121)
# ax_img1.plot(horizontal_profile, '.')
# ax_img1.set_xlabel('Pixel Intensity')
# ax_img1.set_ylabel("Number of pixels")

# # ax_img1.set_xlim([350,len(image_data[0])-1])
```

```python
first_plateau_mean = []
last_plateau_mean = []
min_value = []
pixel_index_of_min_value = []

for _data in image_data:

    _profile = _data[profile_height,:]
    first_plateau_mean.append(np.mean(_profile[first_plateau_range[0]: first_plateau_range[1]]))
    last_plateau_mean.append(np.mean(_profile[last_plateau_range[0]: last_plateau_range[1]]))
    min_value.append(np.min(_profile))
    pixel_index_of_min_value.append(np.argmin(_profile))    

fig4 = plt.figure(figsize=(20, 10))

ax_img1 = plt.subplot(121)
ax_img1.plot(first_plateau_mean, '.', label="first plateau")
ax_img1.plot(last_plateau_mean, 'b.', label='last plateau')
ax_img1.legend()
ax_img1.set_title("mean value of plateau vs image index")
ax_img1.set_xlabel("Image index")
ax_img1.set_ylabel("mean counts value")

ax_img2 = plt.subplot(122)
ax_img2.plot(min_value, "+")
ax_img2.set_title("minimum value")
ax_img2.set_xlabel("Image index")
ax_img2.set_ylabel("Counts")

plt.show()
```

```python
left_side_of_sample_pixel_value = []

for _index, _data in enumerate(image_data):

    profile = _data[profile_height,:]
    left_profile = profile[first_plateau_range[0]: pixel_index_of_min_value[_index]]
    
    mean_value = np.mean([first_plateau_mean[_index], min_value[_index]])

    found_min = False
    for _x_index, _x in enumerate(left_profile):
        
#         print(f"_x:{_x} and mean_value:{mean_value}")
        
        if _x < mean_value:
            left_side_of_sample_pixel_value.append(_x_index)
            found_min = True
            break
    
#     if found_min == False:
#         left_side_of_sample_pixel_value.append(np.nan)
    

fig5 = plt.figure(figsize=(20, 10))

ax_img1 = plt.subplot(121)
ax_img1.plot(left_side_of_sample_pixel_value, '.')
ax_img1.set_xlabel("file index")
ax_img1.set_ylabel("left position of sample")
plt.show()
```

```python
right_side_of_sample_pixel_value = []

for _index, _data in enumerate(image_data):

    profile = _data[profile_height,:]
    right_profile = profile[pixel_index_of_min_value[_index]: last_plateau_range[1]]
    
    mean_value = np.mean([last_plateau_mean[_index], min_value[_index]])

    found_min = False
    for _x_index, _x in enumerate(right_profile[::-1]):
        
#         print(f"_x:{_x} and mean_value:{mean_value}")
        
        if _x < mean_value:
            right_side_of_sample_pixel_value.append(len(profile)-_x_index)
            found_min = True
            break
    
#     if found_min == False:
#         left_side_of_sample_pixel_value.append(np.nan)
    

fig5 = plt.figure(figsize=(20, 10))

ax_img1 = plt.subplot(121)
ax_img1.plot(right_side_of_sample_pixel_value, '.')
ax_img1.set_xlabel("file index")
ax_img1.set_ylabel("right position of sample")
plt.show()
```

## Calculate sample size

```python
width_sample = np.array(right_side_of_sample_pixel_value) - np.array(left_side_of_sample_pixel_value)
```

```python
fig6 = plt.figure(figsize=(20, 10))

ax_img1 = plt.subplot(111)
ax_img1.plot(width_sample, '*')
ax_img1.set_xlabel("file index")
ax_img1.set_ylabel("Width of sample (pixel)")
ax_img1.set_title("Width of sample1 StickB")
plt.show()
```

```python
output_file_name = '/Users/j35/Desktop/stickB_sample1_width.txt'
x_axis = np.arange(len(width_sample))
y_axis = width_sample

data = []
for _x, _y in zip(x_axis, y_axis):
    data.append(f"{_x}, {_y}")
    
metadata = ["#Stick B, sample 1", "#", "#File index, width of sample"]

def make_ascii_file(metadata=[], data=[], output_file_name='', dim='2d', sep=','):
    f = open(output_file_name, 'w')
    for _meta in metadata:
        _line = _meta + "\n"
        f.write(_line)
        
    for _data in data:
        if dim == '2d':
            _str_data = [str(_value) for _value in _data]
            _line = sep.join(_str_data) + "\n"
        else:
            _line = str(_data) + '\n'
        f.write(_line)
       
    f.close()

make_ascii_file(metadata=metadata,
               data=data,
               output_file_name=output_file_name,
               dim='1d')
```
